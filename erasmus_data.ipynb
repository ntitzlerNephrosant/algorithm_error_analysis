## Goal: ##

Find out how sample variance effects the range of QScores produced. We hope to discover the analyte drivers for 'bucket', see if we can create buckets that capture a clear analyte driver, and observe how much variance in each bucket causes the score to shift into a new region or 'bucket' of scores. 

To accomplish this we will test:<br>

```python
def modify_item(l, index, inc, minn, maxx):
    """
    Create new rows based on the inputted row L. returns a nested array 
    containg rows with an incremented value based on the inc and maxx
    
    Parameters
    __________
    
    l: (list)
        A single sample row of any length
        
    index: (int)
        The index too be modified
    
    inc: (int)
        The increment to modify the function with
        
    minn: (int)
        The value that the increment will start at
        
    maxx: (int)
        The item that the increment will generate up too
        
    Returns
    _______
    
    result: (list<list>)
        a nested array containg rows with an incremented value at the infex from min -> maxx based on increment
        
    
    """
    # add a bunch of samples in increments of 10 from 1 - 200. update the ID as you go
    og_sample = l.copy()
    val = minn
    inc = inc
    result = [og_sample]
    
    first = True
    
    while val < maxx:
            
        sample_id = og_sample[0] +"_"+ str(val)
        new_row = []
        new_row = og_sample.copy()
        new_row[0] = sample_id
        new_row[index] = val
        result.append(new_row)
        
        val += inc
        
    
    return result

import math



def df_to_2d(df):
	"""	
	Converts a dataframe to a nest list of arrays. Assumes there is a header column
	
	Parameters
	__________

	df: pd.DataFrame
		a dataframe object


	Returns
	_______

	l: List<list>
		a 2d list where the first row is a header column

	"""

	arr = list(df.values.tolist())
	arr.insert(0, list(df.columns.tolist()))

	return arr

def two_d_to_df(l:list):
	"""	
	Converts a list with headers to a pandas df, where the first row of the list becomes the column for the dataframe


	Parameters
	__________

	l: List<list>
		a 2d list where the first row is a header column


	Returns
	_______

	df: pd.DataFrame
		a dataframe object 

	"""
	return pd.DataFrame(data=l[1:], columns=l[0])


def calculate_total_change(scores:list):
    # For each point calculate the difference between that point and the original
    
    # scores = [original, non-original, non-original]
    original_score = scores.pop(0)
    total_change = 0
    
    for item in scores:
        res = abs(original_score-item)
        total_change += res
    
    return total_change

def calculate_change_prev(scores:list):
    # For each point calculate the difference between that point
    # and the previous point, with the original first point
    
    # scores = [original, non-original, non-original]
    original_score = scores.pop(0)
    total_change = 0
    
    prev_score = original_score
    for item in scores:
        res = abs(prev_score-item)
        total_change += res
        prev_score = item
    
    return total_change

def test_calculate_total_change():
    # index 1 represents the original score
    scores = [50, 40, 40, 40, 40, 40, 40, 40, 40]
    res    = calculate_change_prev(scores)
    assert res==10, "Calculate change prev is not outputting right. Correct is 10. Current output: " + str(res)
    
    scores = [50, 40, 50, 40, 40, 40, 40, 40, 40]
    res    = calculate_change_prev(scores)
    assert res==30, "Calculate change prev is not outputting right. Correct is 10. Current output: " + str(res)
    
def upperAndLower(variance:float, value:float):
		"""


		Parameters
		__________

		range_percent: (float)
			Percent which will be multiplied to 'value' to find upper and lower bound

		Value: (float)
			Value to find variance of


		Returns
		_______
		
		lower: (float)
			result of the min value minus the minValue * the range_percent

		upper: (float)
			result of the min value minus the maxValue * the range_percent

		
		"""
		lower = float(value) - float((value * variance))
		upper = float(value) + float((value * variance))
		return lower, upper

test_calculate_total_change()
```

```python
import operator
class Single_sample_results():
    """
    Holds data and analysis functions/information for a single sample
    
    Parameters
    __________
    
    analyte_results: <list(Analyte_results)>
        List of Analyte_results
    
    key: <str>
        Key to sort self.analyte_results with
    
    meta_info: <dict>
        Meta info that contains 
        
    print("Analyte_name: ",self.analyte_name)
        print("model_name: ",self.model_name)
        print("Original sample: ",self.original_sample)
        print("Total_change_prev: ", self.total_change_prev)
        print("Total_change_prev_div_num_samples: ",self.total_change_prev_div_num_samples)
        print("Variance: ",self.variance)
        print("Error Range: ",self.error_range)
    """
    
    def __init__(self, analyte_results,meta_info, key="total_change_prev"):
        
        # list that holds the Analyte_results classes
        self.analyte_results       = sorted(analyte_results, key=operator.attrgetter("total_change_prev"), reverse=True) # sorted by total_change_prev
        self.a_results_num_buckets = sorted(analyte_results, key=operator.attrgetter("num_bucket_changes"), reverse=True) # sorted by total_change_prev
        
        self.model_name      = meta_info["model_name"]
        self.sample_id       = meta_info["sample_id"]
        self.original_sample = meta_info["original_sample"]
        self.variance        = meta_info["variance"]
        self.headers         = meta_info["headers"]
        self.assay_ranges    = meta_info["assay_ranges"]
        self.size            = meta_info["size"]
        self.sample_sets     = self.get_sample_sets(self.analyte_results)
        self.sample_names    = self.sorted_result_names(self.analyte_results,key="total_change_prev")
        
        self.num_buckets_sample_sets     = self.get_sample_sets(self.a_results_num_buckets)
        self.num_buckets_sample_names    = self.sorted_result_names(self.a_results_num_buckets, key="num_bucket_changes")

        self.key             = key

        
    def get_sample_sets(self, a_results=None):
        """
        Put all sample sets into a list
        """
        
        sets = []
        for a in a_results:
            s = a.sample_set.copy()
            
            sets.append(s)
        return sets
        
        
        
    def sorted_result_names(self,a_results, key="total_change_prev"):
        results = []
        
        
        
        for a in a_results:
            if key == "total_change_prev":
                k_val = str(a.total_change_prev)
            elif key == "num_bucket_changes":
                k_val = str(a.num_bucket_changes)
            else:
                raise ValueError("Key Not recognized")
            s = a.analyte_name +": "+ key + ": " + k_val
            results.append(s)
        return results
        
    
    
    def __repr__(self):
        print("#####################################################")
        print("Sample_ID:     ",self.sample_id)
        print("Sample:        ",self.original_sample)
        print("variance:      ",self.variance)
        print("")
        t = "Sorted Results by total_change_prev:"
        print(t)
        for item in self.sample_names:
            print("     "+item)
            
        print("")
        t = "Sorted Results by num_bucket_changes:"
        print(t)
        for item in self.num_buckets_sample_names:
            print("     "+item)
        print("#####################################################")
        return ""
        
    def plot(self,index=9):
        # Class specific wrapper for multi score plotting
  
        ids       = [self.sample_id] * len(self.sample_sets[0][1:])
        self.plot_scores(data_list=self.sample_sets, index=index, ids=ids, names=self.sample_names)
        
        
    
    def plot_scores(self, data_list, index=9, ids=None, names=None):
        """
        Create plot of datas scores for all analytes in the bucket.
        
        If names exists, indexes are equivalent to plots in data_list
        

        Parameters
        __________

        data_list: (list<array>)
            A list of datasets in 3d array format, where the first row of each 2d sample set array is a header column.

        indexes: int
            Index in 2d of column to graph
            
        ids: <list<str>>
            The a list of ID's that represents the original IDs. If not present, take from the data_list
            item

        """
        
        colors = ["red","orange","#FFE900","green","royalblue","plum","pink","tomato"]
        
        
        plt.figure(figsize=(16, 10))
        plt.autoscale(False, axis='y')

        # Set IDs
        color = "red"
        if ids == None:
            ids = column(data, 0)[1:]

        # axis measurments
        x = list(range(0,len(ids)))
        plt.xticks(x, ids)
        y = np.arange(0, 105, 5)
        plt.yticks(y)

        # Set Scores
        scores  = []
        for data_set in data_list:
            scores.append(column(data_set, index)[1:])
        
        # Set titles
        titles  = []
        if names == None:
            titles.append(data[0][index])
        else:
            for n in names:
                titles.append(n)
                
        # Plot Scores    
        for i in range(len(scores)):
            s = scores[i]
            plt.plot(scores[i], '-o', label=titles[i], color=colors[i])
            

        # Labels, ID, and Score
        plt.legend(loc="upper left")
        plt.xlabel("ID")
        plt.ylabel("Score")

        plt.tick_params(axis="x", which="major", labelsize=8, rotation=90) # adjust axis dimensions
        plt.tick_params(axis="x", which="major", labelsize=8, rotation=90)
        
        # bucket marks
        """
        group_0_16   = []
        group_17_32  = []
        group_33_55  = []
        group_56_80  = []
        group_81_200 = []
        """
        plt.plot([0]*len(x), linestyle="dashed")
        plt.plot([16]*len(x), linestyle="dashed")
        plt.plot([32]*len(x), linestyle="dashed")
        plt.plot([55]*len(x), linestyle="dashed")
        plt.plot([80]*len(x), linestyle="dashed")
        plt.plot([100]*len(x), linestyle="dashed")
        
        plt.grid(axis="x", color="gray",alpha=.5,linestyle=":")
        plt.grid(axis="y", color="gray",alpha=.5,linestyle=":")
        plt.show()
        


    
        
        
        
        
```
